```python
#Q5
from collections import deque

class BinaryTree:
    # ... (previous code)

    def bfs(self):
        if not self.root:
            return []

        result = []
        queue = deque([self.root])

        while queue:
            node = queue.popleft()
            result.append(node.value)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        return result

    def dfs(self):
        if not self.root:
            return []

        result = []
        stack = [self.root]

        while stack:
            node = stack.pop()
            result.append(node.value)

            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)

        return result
```

```python
#Q8
class BinaryTree:
    # ... (previous code)

    def count_subtrees_with_sum(self, node, x):
        self.count = 0

        def subtree_sum(current_node):
            if not current_node:
                return 0

            left_sum = subtree_sum(current_node.left)
            right_sum = subtree_sum(current_node.right)
            total_sum = left_sum + right_sum + current_node.value

            if total_sum == x:
                self.count += 1

            return total_sum

        subtree_sum(node)
        return self.count
```

```python
#Q9
class BinaryTree:
    # ... (previous code)

    def max_level_sum(self, node):
        if not node:
            return 0

        max_sum = float('-inf')
        queue = deque([node])

        while queue:
            level_sum = 0
            level_size = len(queue)

            for _ in range(level_size):
                current_node = queue.popleft()
                level_sum += current_node.value

                if current_node.left:
                    queue.append(current_node.left)
                if current_node.right:
                    queue.append(current_node.right)

            max_sum = max(max_sum, level_sum)

        return max_sum
```
